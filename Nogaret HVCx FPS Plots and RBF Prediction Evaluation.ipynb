# **Nogaret HVCx FPS Plots and RBF Prediction Evaluation**

```python
import numpy as np
import matplotlib.pyplot as plt
from current_voltage_Fourier_power_spectrum_and_plot import produce_IV_Fourier_power_spectrum_and_plot
import warnings; warnings.simplefilter('ignore')

neuron_directory = "Nogaret_HVCx_model_I(t)_and_V(t)"

# Reduce array to only important timesteps to be analyzed. Will be removed from V and I and/or set to nan in
# function produce_IV_Fourier_power_spectrum_and_plot()
remove_indices = False
range_list = [range(1)]

freq_units = "kHz" # frequency units
sampling_rate = 50.0 # kHz (1/ms)
delta_t = 1.0/sampling_rate  # this is milliseconds (1.0/50kHz= 0.02ms)
```

Using Jason's Julia Nogaret HVCx code, I calculated a simulated HVCx V(t) in response to the 700pA L63x currents I gave to Arij (plus one more at time dilation = 4). This means I evaluated for 5 time-stretched variations on the same current. I will check their {I(t),V(t)} Fourier power spectrum (FPS) and the ability to train and predict on the {I(t),V(t)} data using RBFs. I will then analyze whether FPS and RBF prediction success are correlated.

Files (I(t) and V(t)) for this are in folder "Nogaret_HVCx_model_I(t)_and_V(t)/".

## Fourier Power Spectra

### Time dilation = 0.2

```python
# Basic Plotting
save_and_or_display = "display"
neuron_number = "Nogaret-HVCx" # This is an x-projecting neuron
episode_number = "0.2"

# everything in filename except "_voltage.png" for instance. Used as neuron name
a_filename_pretext = "I_L63_x_time_dilation=0.2range=700_(I)_converted_units"

V = np.loadtxt(neuron_directory +"/Nogaret_V(t)_Response_to_I_L63_x_time_dilation=0.2range=300_(I).txt")[:,1]
I_temp = np.loadtxt(neuron_directory +"/I_L63_x_time_dilation=0.2range=700_(I)_converted_units.txt")[:,1]

num_timesteps = V.shape[0]
t_final = num_timesteps * delta_t  # measured in ms
t_arr = np.arange(start=0, stop= t_final, step= delta_t) # units: ms
xlim = (0,0.5)
print("num_timesteps: "+str(num_timesteps))

I = I_temp[:num_timesteps]
# Finally, produce the I(t) and V(t) plots, and their Fourier power spectrum plots:
FPS_list_omission, freq_array_omission = produce_IV_Fourier_power_spectrum_and_plot(V,I,range_list,t_arr,
                                                                                    a_filename_pretext,freq_units,
                                           neuron_name=neuron_number, remove_indices=remove_indices, xlim=xlim,
                                           save_and_or_display = save_and_or_display,
                                           current_units = "pA", voltage_units = "mV")
```
